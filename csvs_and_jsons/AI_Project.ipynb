```python
movie_vector = {}

for index, row in df.iterrows():
    movie_title = row['original_title']

    # Values: List of genre indicators
    genre_values = row[2:-10].tolist()  # Get all genre columns (0s and 1s)
    vote_one_hot = []

    for vote_col in [col for col in row.index if 'vote_average' in col]:  # Get all vote average columns
        # The column name holds the vote number (e.g., 'vote_average_1.0' means a vote of 1)
        count = row[vote_col]

        vote_one_hot.append(1 if count > 0 else 0)

    # Combine genre values with one-hot encoding for votes
    movie_vector[movie_title] = genre_values + vote_one_hot

for title, vector in movie_vector.items():
    print(f"{title}: {vector}")
```

```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Filter movies in df that have vectors
df_filtered = df[df['original_title'].isin(movie_vector.keys())].reset_index(drop=True)

# Convert dictionary to NumPy array for cosine similarity
movie_vectors_array = np.array([movie_vector[title] for title in df_filtered['original_title']])

# Compute cosine similarity
cosine_similarities = cosine_similarity(movie_vectors_array)

# Find the most similar movie
def find_most_similar_movie(liked_movie):
    if liked_movie in df_filtered['original_title'].values:
        liked_movie_index = df_filtered[df_filtered['original_title'] == liked_movie].index[0]

        # Debugging prints
        print(f"Liked movie index: {liked_movie_index}")
        print(f"Cosine similarities size: {cosine_similarities.shape[0]}")

        # Ensure that the liked_movie_index is within bounds
        if liked_movie_index < cosine_similarities.shape[0]:
            similarity_scores = cosine_similarities[liked_movie_index]
            most_similar_index = np.argsort(-similarity_scores)[1]  # Get the most similar (skip self)
            most_similar_movie = df_filtered['original_title'].iloc[most_similar_index]
            return most_similar_movie
        else:
            return f"Error: Index for '{liked_movie}' exceeds cosine similarity array bounds."
    else:
        return f"Movie '{liked_movie}' not found in the dataset."

# Implement DFS for expanding recommendations
def dfs(movie_title, depth=0, max_depth=2, visited=None):
    if visited is None:
        visited = set()
    if depth > max_depth or movie_title in visited:
        return []

    visited.add(movie_title)

    if movie_title not in df_filtered['original_title'].values:
        return []  # Return empty if the movie title is not valid

    recommendations = []
    movie_index = df_filtered[df_filtered['original_title'] == movie_title].index[0]

    # Ensure movie_index is within bounds
    if movie_index < cosine_similarities.shape[0]:
        for j, score in enumerate(cosine_similarities[movie_index]):
            if j != movie_index and score > 0.8:  # Adjust threshold as needed
                recommendations.append(df_filtered['original_title'].iloc[j])
                recommendations.extend(dfs(df_filtered['original_title'].iloc[j], depth + 1, max_depth, visited))

    return recommendations

# Get recommendations based on user choice
def get_recommendations(liked_movie):
    if liked_movie in movie_vector:  # Check if movie has a vector
        most_similar_movie = find_most_similar_movie(liked_movie)
        print(f"The most similar movie to '{liked_movie}' is '{most_similar_movie}'.")

        if "Error" not in most_similar_movie:  # Only proceed if no error occurred
            # Expand recommendations using DFS
            expanded_recommendations = dfs(most_similar_movie, max_depth=2)
            return list(set(expanded_recommendations))  # Remove duplicates
        else:
            return most_similar_movie  # Return error message
    else:
        return f"Cannot recommend based on '{liked_movie}' as it is not in movie_vector."

# Example usage
liked_movie = input("Enter the title of a movie you like: ")
recommendations = get_recommendations(liked_movie)
print(f"Expanded recommendations: {recommendations}")
```
